# Investigating players according to their age

```python
# Showing age distribution of all players
plt.figure(figsize=(10, 6))
sns.histplot(fifa['Age'], bins=10, kde=True)
plt.title('Age distribution of Players')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

So, we can say that the vast majority of players age are between (18- 32)
**bold text**

***Let's see the relation between age and overall rating ***

```python
# scatter plot between age and average overall rating
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Overall', data=fifa)
plt.title('Age vs Overall Rating')
plt.xlabel('Age')
plt.ylabel('Overall Rating')
plt.show()
```

```python
# the same plot using boxplot
plt.figure(figsize=(14, 6))
sns.boxplot(x='Age', y='Overall', data=fifa)
plt.title('Age vs Overall Rating')
plt.xlabel('Age')
plt.ylabel('Overall Rating')
plt.show()
```

So, we can get that players rating tend to get higher rating as they enter their twenties

**But waht about countries, what country has the most of young players?**

```python
# what countries produce the most of youngest players (under 23)
Under23_players = fifa.loc[fifa['Age'] <= 23]
N_Under23_players = Under23_players.groupby(['Nationality'])['Overall'].count().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
plt.bar(N_Under23_players.index, N_Under23_players.values, color='#c180ff')
plt.xlabel('Nationality')
plt.ylabel('Number of Players')
plt.title('Top 10 Countries with the Most player <= Under 23')
plt.xticks(rotation=45)
plt.show()
```

We can see that England is by far the country that produces the most player <=*23*

# Analysing data according to countries

```python
# Defining some important variables
top_countries = fifa['Nationality'].value_counts().head(10)

top_countries_name = top_countries.index
top_countries.index
```

```python
# Showing countries with the most players
top_countries_srt= top_countries.sort_values(ascending=True)
plt.figure(figsize = (12, 6))
x= top_countries_srt.index
y= top_countries_srt.values
plt.bar(x,y,color= '#e33dea')
plt.xlabel('Countries')
plt.ylabel('Number of Players')
plt.title('Top 10 Countries with most number of players')
plt.show()
```

So, England has the nost players followed by Germany

**Starting to analyse the top 10 countries with the most players**

```python
df_country_age = fifa[fifa['Nationality'].isin(top_countries_name)][['Nationality', 'Age']]
# Create the boxplot with informative title
plt.figure(figsize=(12, 6))  # Set appropriate figure size
sns.boxplot(x='Nationality', y='Age', data=df_country_age, palette='rocket')
plt.title('Age Distribution of Players from Top 10 Countries')
plt.xlabel('Nationality')
plt.ylabel('Age')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust spacing for clarity
plt.show()
```

We noticed that France, Germany, England and Netherlands have youngest set of players and Japan, Argentina and Brazil have eldest set of players.
So, how about investigating those four countries to see which country produce the most players

**That was about age what about overall rating? what country produces the best players? which countries has better goalkeepers, back, forword and midfield?**

```python
Top_countries_overall = fifa.loc[fifa['Nationality'].isin(top_countries_name) & fifa['Overall']]
```

```python
avg_ratings = Top_countries_overall.groupby('Nationality')['Overall'].mean()
sorted_data = avg_ratings.sort_values(ascending=True)
plt.figure(figsize=(12, 6))
plt.title('Players average rating in top 10 countries')
plt.bar(sorted_data.index, sorted_data.values, color='#ccccff')
plt.xlabel('Countries')
plt.ylabel('Average Overall Rating')
plt.xticks(rotation=45, ha='right')
plt.show()
```

So Brazil has the most players with highest average rating followed by Spain and Nertherlands

**which country has the best GoalKeepers**


```python
Top_countries_GK= Top_countries_overall.loc[Top_countries_overall['Preferred Positions Type'] == 'GoalKeeper']
```

```python
# showing average rated goalkeepers in top countries
GK = (Top_countries_GK.groupby('Nationality')['Overall'].mean())
sorted_GK = GK.sort_values(ascending=True)
plt.figure(figsize=(12, 6))
plt.title('Top Countries with Best GK Ratings')
plt.bar(sorted_GK.index, sorted_GK.values, color='#ffc34d')
plt.xlabel('Countries')
plt.xticks(rotation=45,ha= "right")
plt.ylabel('Average Overall Rating')
plt.show()
```

```python
from math import pi
# Define the data
categories = ['Passing', 'Shooting', 'Dribbling', 'Defending', 'Physical']
player1 = [75, 85, 90, 70, 80]  # Example values for Player 1
player2 = [80, 70, 85, 75, 90]  # Example values for Player 2

# Number of variables
N = len(categories)

# What will be the angle of each axis in the plot
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]

# The plot is made in a circular (not polygon) space, so we need to "complete the loop"
player1 += player1[:1]
player2 += player2[:1]

# Initialize the radar plot
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

# Draw one axe per variable and add labels
plt.xticks(angles[:-1], categories)

# Draw ylabels
ax.set_rlabel_position(0)
plt.yticks([20, 40, 60, 80, 100], ["20", "40", "60", "80", "100"], color="grey", size=7)
plt.ylim(0, 100)

# Plot data for Player 1
ax.plot(angles, player1, linewidth=1, linestyle='solid', label='Player 1')
ax.fill(angles, player1, 'b', alpha=0.1)

# Plot data for Player 2
ax.plot(angles, player2, linewidth=1, linestyle='solid', label='Player 2')
ax.fill(angles, player2, 'r', alpha=0.1)

# Add a legend
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

# Show the plot
plt.show()
```
